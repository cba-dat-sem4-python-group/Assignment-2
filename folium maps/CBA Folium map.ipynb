```python
# install folium with ( $ pip install folium )

import folium

# CPHBusiness locations
locations = [
    { 'description': 'CPHBusiness, Lyngby', 'coordinates': [55.7702835,12.509539] },
    { 'description': 'CPHBusiness, Hillerød', 'coordinates': [55.919916, 12.2799283] },
    { 'description': 'CPHBusiness, Nørrebro', 'coordinates': [55.686164, 12.5567113] },
    { 'description': 'CPHBusiness, Søerne', 'coordinates': [55.682001,12.5604232] },
    { 'description': 'CPHBusiness, City', 'coordinates': [55.6820535, 12.5740235] },
    { 'description': 'CPHBusiness, Rønne', 'coordinates': [55.106774,14.7105143] },
]

# creating a folium map
geomap = folium.Map (
    
    # start target (copenhagen)
    location=[55.6713442, 12.5237849],
    
    # zoom amount, zealand in focus
    zoom_start = 9
)

# adds markers with locations to the folium map
for location in locations:
    
    # creates marker
    folium.Marker(
        
        # location longitude and langitude
        location['coordinates'],
        
        # info in popup box (html)
        popup = (
            '<b>' + location['description'] + '</b> <br>' 
            + str(location['coordinates'][0]) + '<br>' 
            + str(location['coordinates'][1]))
    
    # adds marker to the folium map
    ).add_to(geomap)

# shows the folium map
geomap

# note:: ( CPHBusiness, Rønne ) is outside the default target zone
```
